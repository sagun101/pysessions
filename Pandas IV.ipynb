```python
locations = ["A", "B", "C", "D", "E", "F"]
data = [1, 8, 3, 7, 9, 2.0]
```

```python
idx = pd.date_range('2016-06-01', periods=6, freq='MS')
```

```python
df1 = pd.DataFrame(data, index=idx, columns=["Revenue"])
```

```python
df1
```

```python
df1["locations"] = locations
```

```python
df2 = pd.DataFrame([10.0, 10, 9, 8, 7, 5, 6, 6, 5, 8],
                   index=pd.date_range('2016-06-01', periods=10, freq='MS'),
                   columns=["Revenue"])
```

```python
df2
```

```python
df2["locations"] = locations + ["G", "H", "I", "J"]
```

```python
df2
```

```python
df = pd.concat([df1, df2])
```

```python
df
```

```python
ndf = df.copy()
```

```python
ndf['Revenue'].mean()
```
